```python
import numpy as np

arr = np.arange(1, 26).reshape(5,5)

print(arr)
```

```python
center = arr[0:4,0:4]

print(center)
```

```python
greater_than_15 = arr[arr>15]
print(greater_than_15)
```

```python
arr_even_replace = arr.copy()
arr_even_replace[ arr_even_replace % 2 ==0] = 0
print(arr_even_replace)
```

```python
diagonal_elements = np.diag(arr)
print("\nDiagonal elements:\n", diagonal_elements)
```
