```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)

plt.xlabel("Numbers")
plt.ylabel("Squares")
plt.title("Square Numbers")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y, color='red', linestyle='--',marker='o')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]

y1 = [2,4,6,8]
y2 = [1,3,5,7]
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]

y1 = [2,4,6,8]
y2 = [1,3,5,7]

plt.plot(x, y1, label="Even Pattern")
plt.plot(x, y2, label="Odd Pattern")

plt.legend()

plt.show()
```
